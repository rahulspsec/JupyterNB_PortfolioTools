```python
from portfoliotools.screener.utility.util import getHistoricStockPrices
import pandas as pd
stock_list = ['SBI', 'HDFC', 'INFY']
prices = pd.DataFrame()
for stock in stock_list:
    prices[stock] = getHistoricStockPrices(stock, days = 1000)['Adj Close']

prices.dropna(inplace = True)
prices
```
